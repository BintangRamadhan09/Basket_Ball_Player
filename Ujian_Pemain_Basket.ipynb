### 1. NBA Player

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

#### Collecting datasets


```python
player = pd.read_csv('all_seasons.csv')
player
```

```python
player.columns
```

```python
# Checking missing values
player.isnull().sum()
```

```python
player_nba = player[['player_name','age','player_height','player_weight','pts','reb']]
```

```python
player_nba.head()
```

```python
player_nba['target_player'] = player_nba.apply(lambda x: 1 if (x['age'] <=25) & 
                                               (x['player_height'] >= 180) & 
                                               (x['player_weight'] <= 90) & 
                                               (x['pts']>=6) &
                                               (x['reb']>=3) else 0, axis=1)
```

```python
player_nba.tail()
```

```python
player_nba[player_nba['target_player'] == 1]
```

### Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
stdScaler = StandardScaler()
Std = stdScaler.fit_transform(player_nba[['age','player_height','player_weight','pts','reb']])
dfStd = pd.DataFrame(Std,columns=['age','player_height','player_weight','pts','reb'])
dfStd
```

### Splitting Datasets (82% Train)

```python
from sklearn.model_selection import train_test_split
x = player_nba[['age','player_height','player_weight','pts','reb']]
y = player_nba['target_player']
xtr,xts,ytr,yts = train_test_split(x,y,test_size=.82,random_state=100)
```

```python
xtr.shape
```

### Modelling

```python
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, roc_curve
```

#### Using Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
modelRF = RandomForestClassifier()
print(modelRF.fit(xtr,ytr))
# modelRF.score(xts,yts)
```

```python
# Using Hyperparameter Tuning

# Creating Parameters
n_estimators = [10,100,1000]
criterion = ['gini']
max_depth = [10,20,30,40,'none']
param = {'n_estimators':n_estimators,'criterion':criterion,'max_depth':max_depth}
```

```python
from sklearn.model_selection import GridSearchCV
modelCV =GridSearchCV(modelRF,param, cv=5)
```

#### Deciding the best parameters


```python
modelCV.fit(xtr,ytr)
modelCV.best_params_
```

#### New Model after using hyperparameter tuning

```python
modelRFbaru = RandomForestClassifier(criterion='gini',max_depth=40,n_estimators=10)
modelRFbaru.fit(xtr,ytr)
modelRFbaru.score(xts,yts)
```

<hr>

#### Using Decision Tree Classifier

```python
from sklearn import tree
modelTree = tree.DecisionTreeClassifier()
print(modelTree.fit(xtr,ytr))
modelTree.score(xts,yts)
```

```python
# Creating Parameters
splitter = ['best','random']
criterion = ['gini','entropy']
max_depth = [10,20,30,40,'none']
param = {'splitter':splitter,'criterion':criterion,'max_depth':max_depth}
```

```python
modelCVTree =GridSearchCV(modelTree,param, cv=5)
```

```python
# deciding the best parameters
modelCVTree.fit(xtr,ytr)
modelCVTree.best_params_
```

```python
# model after hyperparameter tuning
modelTreeNew = tree.DecisionTreeClassifier(criterion='gini',max_depth=10,splitter='best')
modelTreeNew.fit(xtr,ytr)
modelTreeNew.score(xts,yts)
```

<hr>

### Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
modelLR = LogisticRegression(solver='liblinear')
print(modelLR.fit(xtr,ytr))
modelLR.score(xts,yts)
```

```python
# Creating Parameters
penalty = ['l1','l2','elasticnet','none']
solver = ['newton-cg','lbfgs','liblinear','sag','saga']
max_iter = [1,10,100,1000,10000]
paramLR = {'penalty':penalty,'solver':solver,'max_iter':max_iter}
```

```python
modelCVLR =GridSearchCV(modelLR,paramLR, cv=5)
```

```python
# deciding the best parameters
modelCVLR.fit(xtr,ytr)
modelCVLR.best_params_
```

```python
# model after hyperparameter tuning
modelLRNew = LogisticRegression(max_iter=10,penalty='l2',solver='liblinear')
modelLRNew.fit(xtr,ytr)
modelLRNew.score(xts,yts)
```

```python
test = player_nba[['age','player_height','player_weight','pts','reb',]]
```

```python
# Prediction
player_nba['prediksi1'] = modelRFbaru.predict(test) 
player_nba['prediksi2'] = modelTreeNew.predict(test)
player_nba['prediksi3'] = modelLRNew.predict(test)
```

```python
player_nba.head()
player_nba
```

### Evaluation Metrics


```python
#  accuracy, precision, recall, F1 score & ROC AUC score
```

```python
from sklearn.metrics import confusion_matrix
```

#### Balanced Accuracy

```python
from sklearn.metrics import balanced_accuracy_score
bas = balanced_accuracy_score
print('accuracy_score:',bas(player_nba['target_player'],player_nba['prediksi1']))
print('accuracy_score:',bas(player_nba['target_player'],player_nba['prediksi2']))
print('accuracy_score:',bas(player_nba['target_player'],player_nba['prediksi3']))
```

#### Recall

```python
from sklearn.metrics import recall_score
rs = recall_score
print('Recall Score(+):',rs(player_nba['target_player'],player_nba['prediksi1'],pos_label=1))
print('Recall Score(+):',rs(player_nba['target_player'],player_nba['prediksi2'],pos_label=1))
print('Recall Score(+):',rs(player_nba['target_player'],player_nba['prediksi3'],pos_label=1))
print('Recall Score(-):',rs(player_nba['target_player'],player_nba['prediksi1'],pos_label=0))
print('Recall Score(-):',rs(player_nba['target_player'],player_nba['prediksi2'],pos_label=0))
print('Recall Score(-):',rs(player_nba['target_player'],player_nba['prediksi3'],pos_label=0))
```

#### Precision

```python
from sklearn.metrics import precision_score
ps= precision_score
# Precision(+)
print('Precision(+):',ps(player_nba['target_player'],player_nba['prediksi1'],pos_label=1))
print('Precision(+):',ps(player_nba['target_player'],player_nba['prediksi2'],pos_label=1))
print('Precision(+):',ps(player_nba['target_player'],player_nba['prediksi3'],pos_label=1))
# Precision (-)
print('Precision(+):',ps(player_nba['target_player'],player_nba['prediksi1'],pos_label=0))
print('Precision(+):',ps(player_nba['target_player'],player_nba['prediksi2'],pos_label=0))
print('Precision(+):',ps(player_nba['target_player'],player_nba['prediksi3'],pos_label=0))
```

#### F1 Score

```python
from sklearn.metrics import f1_score
# F1 Score(+)
print('F1 Score(+):',f1_score(player_nba['target_player'],player_nba['prediksi1'],pos_label=1))
print('F1 Score(+):',f1_score(player_nba['target_player'],player_nba['prediksi1'],pos_label=1))
print('F1 Score(+):',f1_score(player_nba['target_player'],player_nba['prediksi1'],pos_label=1))
# F1 Score (-)
print('F1 Score(+):',f1_score(player_nba['target_player'],player_nba['prediksi1'],pos_label=0))
print('F1 Score(+):',f1_score(player_nba['target_player'],player_nba['prediksi2'],pos_label=0))
print('F1 Score(+):',f1_score(player_nba['target_player'],player_nba['prediksi3'],pos_label=0))
```

#### ROC AUC Score

```python
from sklearn.metrics import roc_auc_score
print('ROC AUC Score:',roc_auc_score(player_nba['target_player'],player_nba['prediksi1']))
print('ROC AUC Score:',roc_auc_score(player_nba['target_player'],player_nba['prediksi2']))
print('ROC AUC Score:',roc_auc_score(player_nba['target_player'],player_nba['prediksi3']))
```
